```python
from scipy import stats
yeo_johnson = lambda x: stats.yeojohnson(x)[0]
zscore = lambda x : stats.zscore(x)

df_input.iloc[:,11:]=df_input.iloc[:,11:].apply(yeo_johnson)
df_input.to_csv('MyBrCa_clinical_immunescores_YJtransformed.tsv', sep='\t', index=False)
df_input.iloc[:,11:]=df_input.iloc[:,11:].apply(zscore)
df_input.to_csv('MyBrCa_clinical_immunescores_YJZtransformed.tsv', sep='\t', index=False)
```
